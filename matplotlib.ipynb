MatPlotLib tutorial

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100);
y = np.sin(x)
z = np.cos(x)
print(z)
```

```python
#sine wave
plt.plot(x,z)
plt.show()
```

Adding titles

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
```

```python
y = x**2
plt.plot(x, y)
```

```python
plt.plot(x, y, 'b*')
```

```python
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Latin', 'Hindi']
people = [100, 50, 150, 50, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Latin', 'Hindi']
people = [100, 50, 150, 50, 70]
ax.pie(people, labels = languages, autopct='%1.1f%%')
```

scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.sin(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, y, color = 'g')
ax.scatter(x, z, color = 'b')
```

3D scatter plot


```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z, c=z, cmap='Blues')
```
